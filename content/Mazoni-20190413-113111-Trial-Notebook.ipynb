```python
import numpy as np
import matplotlib as mpl
print(mpl.get_configdir())
print(mpl.matplotlib_fname())
```

```python
%matplotlib inline
import matplotlib.pylab as plt
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 20}

plt.rc('font', **font)
plt.rcParams['figure.figsize'] = 8, 6
```

```python
x = np.arange(0, 5, 0.1);
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('time [s]')
plt.ylabel('signal [mV]')
plt.show()
```
